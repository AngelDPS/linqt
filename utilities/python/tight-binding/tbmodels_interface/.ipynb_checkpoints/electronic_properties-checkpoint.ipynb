```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import kwant
from kwant.builder import HoppingKind as hop
from tight_binding.visualization import compute_band_structure,fermi_surface_2D,spin_texture_2D
import matplotlib.tri as tri #tricontourf
from kwant.builder import HoppingKind as hop
from kwant import wraparound
from tight_binding.wannier_tools import wannier2kwant
from tight_binding.visualization import compute_band_structure,fermi_surface_2D,spin_texture_2D
```

```python
ham_file ="MoTe2_1L.Ham"
geo_file ="MoTe2_1L.Geo"
syst = wannier2kwant( ham_file, xyz_file )
syst = wraparound.wraparound(syst).finalized();
broadening = 0.1
fermi_level=7.4663;
fermi_energy=fermi_level+1.0;

#Compute band structures
nump = 100;
band_paths = [ ("G",1,[0,0,0]),("X",nump,[0.5,0,0]) ,("M",nump,[0.5,0.5,0]),("Y",nump,[0,0.5,0]),("G",nump,[0,0,0]),("M",nump,[0.5,0.5,0]) ]
kvec,bands,path_end = compute_band_structure(syst, band_paths );

numkp = ( 100 , 100  );
fermi_surf_points = fermi_surface_2D(syst,numkp, fermi_energy=fermi_energy, broadening =broadening,do_plot=False);

numkp = ( 100 , 100  );
spin_text_points = spin_texture_2D(syst,numkp, fermi_energy=fermi_energy, broadening =broadening,do_plot=False );
 
```

```python
#Plot spin textures:
X,Y,W =np.transpose( fermi_surf_points );
X,Y,SX,SY,SZ = np.transpose(spin_text_points);
fig, ax = plt.subplots()
ax.set_title("Spin Textures")
spin_texture =ax.quiver(X, Y, SX,SY,SZ, scale=10., cmap = "RdBu" )
energy_contours = ax.tricontourf(X, Y, W, np.arange(0,1.0, .01),extend='both',antialiased=True, cmap = "Greys" , LEVE)
fig.colorbar(energy_contours, ax=ax)
qk = ax.quiverkey(spin_texture, 0.60, 0.85, 1,"max", labelpos='E', coordinates='figure')
fig.colorbar(spin_texture, ax=ax)
plt.show()
```
